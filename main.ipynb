```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")  # Apply theme
sns.boxplot(x="species", y="sepal_length", data=sns.load_dataset("iris"))
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Set figure size
sns.histplot(sns.load_dataset("penguins")["body_mass_g"], bins=30, kde=True)
plt.title("Penguin Body Mass Distribution", fontsize=15)
plt.xlabel("Body Mass (g)")
plt.ylabel("Count")
plt.show()
```

```python
sns.set_palette("pastel")  # Choose color palette
sns.barplot(x="day", y="total_bill", data=sns.load_dataset("tips"))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("iris")

# Create a pair plot
sns.pairplot(df, hue="species", diag_kind="kde")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("penguins").dropna()

# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.violinplot(x="species", y="sepal_length", data=sns.load_dataset("iris"))
plt.show()
```
